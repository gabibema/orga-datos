* Generar un nuevo DataFrame **dolar_maxmin_mes** que contenga la cotización más alta y la mas baja de cada mes. El DF debe tener las siguientes columnas: **año,mes,venta_max_mes,venta_min_mes,compra_max_mes, compra_min_mes**

```python
ds_dollar['fecha']=pd.to_datetime(ds_dollar['fecha'])
```

```python
ds_dollar['dia'] = pd.DatetimeIndex(ds_dollar['fecha']).day
ds_dollar['mes'] = pd.DatetimeIndex(ds_dollar['fecha']).month
ds_dollar['año'] = pd.DatetimeIndex(ds_dollar['fecha']).year

```

```python
ds_meses = pd.DataFrame(data = ds_dollar.groupby(['año','mes']).mean(), columns = ['compra','venta'])
ds_meses.rename(columns = {'compra':'compra_prom_mes', 'venta':'venta_prom_mes'}, inplace = True)
ds_meses
```

```python
dolar_max_min_mes =ds_dollar.groupby(['año','mes']).min()[['venta', 'compra']]
max = ds_dollar.groupby(['año','mes']).max()[['venta', 'compra']]
dolar_max_min_mes.rename(columns = {'compra':'compra_min_mes', 'venta':'venta_min_mes'}, inplace = True)
dolar_max_min_mes['venta_max_mes'] = max.venta
dolar_max_min_mes['compra_max_mes'] = max.compra
dolar_max_min_mes
```

```python

dolar_max_min_mes = ds_dollar.groupby(['año','mes']).agg(
    venta_max_mes=pd.NamedAgg(column="venta", aggfunc="max"),
    venta_min_mes=pd.NamedAgg(column="venta", aggfunc="min"),
    compra_max_mes=pd.NamedAgg(column="compra", aggfunc="max"),
    compra_min_mes=pd.NamedAgg(column="compra", aggfunc="min")
).copy()
dolar_max_min_mes
```

```python
#* Generar un nuevo DataFrame **dolar_hist** que contenga las cotizaciones diarias junto con la máxima, la mínima y la promedio del mes. 
#El DF debe tener las  siguientes columas: **año, mes, dia, compra, compra_prom,# compra_max_mes, compra_min_mes, venta, venta_prom_mes, venta_max_mes, venta_min_mes**
dolar_max_min_mes = ds_dollar.groupby(['año','mes','dia']).agg(
    compra = pd.NamedAgg(column="compra",aggfunc="max"),
    compra_prom = pd.NamedAgg(column="compra", aggfunc="mean"),
    venta_max_mes=pd.NamedAgg(column="venta", aggfunc="max"),
    venta = pd.NamedAgg(column="venta",aggfunc="max"),
    venta_min_mes=pd.NamedAgg(column="venta", aggfunc="min"),
    compra_max_mes=pd.NamedAgg(column="compra", aggfunc="max"),
    compra_min_mes=pd.NamedAgg(column="compra", aggfunc="min")
).copy()
dolar_max_min_mes
```
